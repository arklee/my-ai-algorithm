```python
from sympy import symbols, diff
from sympy.plotting import plot3d
```

```python
x,y = symbols('x,y')
```

```python
z = x**2+y**2-7*x-6*y+x*y+40
```

```python
dzx = diff(z,x)
dzy = diff(z,y)
```

```python
plot3d(z,(x,0,5),(y,0,5))
```

```python
i=1
tx=10
ty=10
while(i<100):
    t0=tx
    t1=ty
    tx=t0-0.1*dzx.evalf(subs={x:t0,y:t1})
    ty=t1-0.1*dzy.evalf(subs={x:t0,y:t1})
    print(str(i)+":"+str(tx)+","+str(ty)+"\n")
    i=i+1
```
